## Figure 1: Overlay of TestHidden_005 image and label

This figure shows the overlay of the image `data/SOTA/TestHidden_005.bmp` with its ground-truth label `data/SOTA/labels/TestHidden_005.tif`, rendered using the tnia overlay helper.

```python
# Generate Figure 1 overlay for TestHidden_005
import os
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
from tnia.plotting.plt_helper import mask_overlay, mask_outline_overlay

# Paths relative to this notebook
image_path = r"../data/SOTA/TestHidden_005.bmp"
label_path = r"../data/SOTA/TestHidden_005.tif"

print("Image:", image_path)
print("Label:", label_path)

# Load image
img = io.imread(image_path)

# Load label (fallback to zeros if not found)
label = io.imread(label_path)

overlay = mask_overlay(img, label)

plt.figure(figsize=(8, 8))
plt.imshow(overlay)
plt.title("Figure 1: TestHidden_005 Overlay")
plt.axis('off')
plt.show()
```

```python
outline_overlay = mask_outline_overlay(img, label)

plt.figure(figsize=(8, 8))
plt.imshow(outline_overlay)
plt.title("Figure 1: TestHidden_005 Outline Overlay")
```
