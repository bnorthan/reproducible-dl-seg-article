## Figure 1: Overlay of TestHidden_005 image and label

This figure shows the overlay of the image `data/SOTA/TestHidden_005.bmp` with its ground-truth label `data/SOTA/labels/TestHidden_005.tif`, rendered using the tnia overlay helper.

```python
from tnia.plotting.plt_helper import mask_overlay
```

```python
# Generate Figure 1 overlay for two images side by side
import os
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
from tnia.plotting.plt_helper import mask_overlay, mask_outline_overlay

# Paths relative to this notebook
image_path = r"../data/SOTA/TestHidden_005.bmp"
label_path = r"../data/SOTA/annotations/class_0/TestHidden_005.tif"

image_path_2 = r"..\\data\\red_blood_cell_neurips\\cell_00193.png"
label_path_2 = r"..\\data\\red_blood_cell_neurips\\annotations\\cell_00193.tiff"

print("Image 1:", image_path)
print("Label 1:", label_path)
print("Image 2:", image_path_2)
print("Label 2:", label_path_2)

# Load images and labels
img1 = io.imread(image_path)
label1 = io.imread(label_path)

img2 = io.imread(image_path_2)
label2 = io.imread(label_path_2)

# Build overlays
overlay1 = mask_overlay(img1, label1)
overlay2 = mask_overlay(img2, label2)

# Show overlays side by side (1 row x 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(overlay1)
axes[0].set_title("Figure 1A: TestHidden_005 Overlay")
axes[0].axis('off')

axes[1].imshow(overlay2)
axes[1].set_title("Figure 1B: cell_00193 Overlay")
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Outline overlays for both images (1 row x 2 columns)
outline_overlay1 = mask_outline_overlay(img1, label1, color = [100, 255, 0], thickness=5)
outline_overlay2 = mask_outline_overlay(img2, label2, color = [100, 255, 0], thickness=5)

fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(outline_overlay1)
axes[0].set_title("Figure 1A: TestHidden_005 Outline Overlay")
axes[0].axis('off')

axes[1].imshow(outline_overlay2)
axes[1].set_title("Figure 1B: cell_00193 Outline Overlay")
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
